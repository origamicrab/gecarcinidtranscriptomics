```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("DCGL_contig_counts.csv")
df.shape
```

```python
df.describe()
```

```python
#transpose
df = df.T
```

```python
df.head()
```

```python
#make first row the column names and then drop first row
df.columns = df.iloc[0]
df = df.iloc[1: , :]
```

```python
df.reset_index(inplace=True)
```

```python
df.rename(columns={"Unnamed:0":"","index":"sample"}, inplace=True)
```

```python
df
```

```python
#melt this into a different form
df_melt = pd.melt(df, id_vars="sample")
df_melt.head(25)
```

```python
meta = pd.read_csv("dc_samples_GL.csv")
```

```python
meta.head()
```

```python
dc_df = pd.merge(df_melt, meta[["sample", "time_point"]], on="sample")
```

```python
dc_df.rename(columns={"Unnamed: 0": "transcript"}, inplace=True)
```

```python
dc_df
```

```python
#groupby transcript then by tp then .sum vales in each category?() ?

aggs = dc_df[dc_df["value"] >= 3].groupby(["transcript", "time_point"])[["value"]].count()
```

```python
aggs
```

```python
aggs.reset_index(inplace=True) #add metadata to each row
aggs.head(10)
```

```python
#drop any rows that have values <3
tri_rep_transcripts = aggs[aggs["value"] >= 3] 
tri_rep_transcripts
```

```python
DCGL_kept = tri_rep_transcripts["transcript"].unique().tolist()
```

```python
len(DCGL_kept)
```

```python
#save as csv
np.savetxt("DCGL_kept.csv", DCGL_kept, delimiter=",", fmt='%s')
```
